# LNDL Parser - Lion Directive Language Parsing

LNDL (Lion Directive Language) is a markup language for LLM-to-system communication. The parser extracts structured outputs from AI responses using XML-like tags and Python-style syntax.

**Core Features:**
- **Lvar Tags**: Variable declarations `<lvar name>value</lvar>`
- **Lact Tags**: Action/tool calls `<lact name>function(...)</lact>`
- **OUT{} Block**: Structured output specification
- **Namespace Support**: Prefixed variables `<lvar Model.field alias>value</lvar>`
- **Balanced Parsing**: Handles nested braces and quoted strings
- **Action Execution**: Lazy execution - only referenced actions run

````python
# Import new Lexer/Parser/AST API
from lionpride.lndl import (
    LactMetadata,
    Lexer,
    LvarMetadata,
    MissingOutBlockError,
    Parser,
    RLvarMetadata,
)
from lionpride.lndl.ast import Lvar, RLvar

# Compatibility helpers for old notebook cells
# These provide the old API using the new Lexer/Parser/AST implementation


def extract_lvars(text: str) -> dict[str, str]:
    """Legacy: Extract raw lvars (non-namespaced)."""
    lexer = Lexer(text)
    tokens = lexer.tokenize()
    parser = Parser(tokens, source_text=text)
    program = parser.parse()

    return {lvar.alias: lvar.content for lvar in program.lvars if isinstance(lvar, RLvar)}


def extract_lvars_prefixed(text: str) -> dict[str, LvarMetadata | RLvarMetadata]:
    """Extract namespaced lvars with metadata."""
    lexer = Lexer(text)
    tokens = lexer.tokenize()
    parser = Parser(tokens, source_text=text)
    program = parser.parse()

    lvars = {}
    for lvar in program.lvars:
        if isinstance(lvar, Lvar):
            lvars[lvar.alias] = LvarMetadata(
                model=lvar.model,
                field=lvar.field,
                local_name=lvar.alias,
                value=lvar.content,
            )
        else:  # RLvar
            lvars[lvar.alias] = RLvarMetadata(
                local_name=lvar.alias,
                value=lvar.content,
            )
    return lvars


def extract_lacts(text: str) -> dict[str, str]:
    """Legacy: Extract raw lacts."""
    lexer = Lexer(text)
    tokens = lexer.tokenize()
    parser = Parser(tokens, source_text=text)
    program = parser.parse()

    return {lact.alias: lact.call for lact in program.lacts if lact.model is None}


def extract_lacts_prefixed(text: str) -> dict[str, LactMetadata]:
    """Extract namespaced lacts with metadata."""
    lexer = Lexer(text)
    tokens = lexer.tokenize()
    parser = Parser(tokens, source_text=text)
    program = parser.parse()

    return {
        lact.alias: LactMetadata(
            model=lact.model,
            field=lact.field,
            local_name=lact.alias,
            call=lact.call,
        )
        for lact in program.lacts
    }


def extract_out_block(text: str) -> str:
    """Extract OUT{} block content."""
    lexer = Lexer(text)
    tokens = lexer.tokenize()
    parser = Parser(tokens, source_text=text)
    program = parser.parse()

    if not program.out_block:
        raise MissingOutBlockError("No OUT{} block found in response")

    # Return formatted fields
    lines = []
    for field_name, value in program.out_block.fields.items():
        if isinstance(value, list):
            lines.append(f"{field_name}: [{', '.join(value)}]")
        elif isinstance(value, str):
            lines.append(f"{field_name}: {value!r}")
        else:
            lines.append(f"{field_name}: {value}")
    return ",\n    ".join(lines)


def parse_out_block_array(content: str) -> dict[str, list[str] | str]:
    """Parse OUT{} block content."""
    text = content.strip()
    if not text.startswith("OUT{"):
        text = f"OUT{{{text}}}"

    lexer = Lexer(text)
    tokens = lexer.tokenize()
    parser = Parser(tokens, source_text=text)
    program = parser.parse()

    if not program.out_block:
        return {}

    # Wrap single identifiers in lists for consistency
    result = {}
    for field_name, value in program.out_block.fields.items():
        if isinstance(value, list):
            result[field_name] = value
        elif isinstance(value, str):
            # Literal string - keep as-is with quotes
            result[field_name] = f'"{value}"'
        elif isinstance(value, (int, float, bool)):
            # Convert to string representation
            result[field_name] = str(value).lower() if isinstance(value, bool) else str(value)
        else:
            result[field_name] = str(value)
    return result


def parse_value(value_str):
    """Parse string value to Python object."""
    if not isinstance(value_str, str):
        return value_str
    value_str = value_str.strip()
    if value_str.lower() == "true":
        return True
    if value_str.lower() == "false":
        return False
    if value_str.lower() == "null":
        return None
    try:
        import ast

        return ast.literal_eval(value_str)
    except (ValueError, SyntaxError):
        return value_str
````

## 1. Extracting Lvars (Legacy Format)

Legacy format uses simple names: `<lvar name>content</lvar>`

````python
# Simple lvar extraction
response = """
<lvar title>Introduction to AI</lvar>
<lvar summary>This is a comprehensive guide to artificial intelligence covering neural networks, machine learning, and deep learning.</lvar>
<lvar author>Dr. Smith</lvar>
"""

lvars = extract_lvars(response)
print(f"Extracted {len(lvars)} lvars:")
for name, value in lvars.items():
    print(f"  {name}: {value[:50]}..." if len(value) > 50 else f"  {name}: {value}")
````

````python
# Multiline content preserved
multiline = """
<lvar code>
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
</lvar>
"""

lvars = extract_lvars(multiline)
print("Multiline lvar content:")
print(lvars["code"])
````

## 2. Extracting Lvars (Prefixed Format)

Prefixed format supports namespacing: `<lvar Model.field alias>value</lvar>`

````python
# Namespaced lvars with aliases
response = """
<lvar Report.title t>Quarterly Financial Report</lvar>
<lvar Report.summary s>Revenue increased by 15% year-over-year.</lvar>
<lvar Report.author>John Doe</lvar>
"""

lvars = extract_lvars_prefixed(response)
print(f"Extracted {len(lvars)} prefixed lvars:")
for local_name, metadata in lvars.items():
    print(f"  {local_name}: {metadata.model}.{metadata.field} = {metadata.value[:40]}...")
````

````python
# Default to field name when no alias provided
response = "<lvar Report.author>John Doe</lvar>"
lvars = extract_lvars_prefixed(response)
metadata = lvars["author"]

print(f"Local name: {metadata.local_name}")  # Uses field name as default
print(f"Model: {metadata.model}")
print(f"Field: {metadata.field}")
print(f"Value: {metadata.value}")
````

## 3. Extracting Lacts (Legacy Format)

Legacy format for action/tool declarations: `<lact name>function(...)</lact>`

````python
# Simple action declarations
response = """
<lact search>search(query="artificial intelligence", limit=10)</lact>
<lact validate>validate_email(email="user@example.com")</lact>
<lact compute>calculate_sum(numbers=[1, 2, 3, 4, 5])</lact>
"""

lacts = extract_lacts(response)
print(f"Extracted {len(lacts)} actions:")
for name, call_str in lacts.items():
    print(f"  {name}: {call_str}")
````

## 4. Extracting Lacts (Prefixed Format)

Prefixed format with namespace support and metadata: `<lact Model.field alias>function(...)</lact>`

````python
# Namespaced actions
response = """
<lact Report.summary s>generate_summary(text="Long article...", max_words=100)</lact>
<lact Report.keywords k>extract_keywords(text="Article content", count=5)</lact>
<lact validate>validate_data(data={"key": "value"})</lact>
"""

lacts = extract_lacts_prefixed(response)
print(f"Extracted {len(lacts)} prefixed actions:")
for local_name, metadata in lacts.items():
    model_field = f"{metadata.model}.{metadata.field}" if metadata.model else "(direct)"
    print(f"  {local_name}: {model_field}")
    print(f"    Call: {metadata.call[:60]}...")
````

````python
# Direct action (no namespace)
response = '<lact search>search(query="AI", limit=5)</lact>'
lacts = extract_lacts_prefixed(response)
metadata = lacts["search"]

print(f"Local name: {metadata.local_name}")
print(f"Model: {metadata.model}")  # None for direct actions
print(f"Field: {metadata.field}")  # None for direct actions
print(f"Call: {metadata.call}")
````

````python
# Warning for Python reserved keywords
import warnings

response = '<lact print>print("hello")</lact>'

with warnings.catch_warnings(record=True) as w:
    warnings.simplefilter("always")
    lacts = extract_lacts_prefixed(response)

    if w:
        print(f"Warning: {w[0].message}")
    print(f"Action extracted: {list(lacts.keys())}")
    print("(Works in LNDL despite being a Python builtin)")
````

## 5. Extracting OUT{} Blocks

OUT blocks specify structured output. Parser handles balanced braces and string escaping.

````python
# Simple OUT block
response = """
Here's the analysis:

OUT{
    title: report_title,
    summary: report_summary
}
"""

out_content = extract_out_block(response)
print("Extracted OUT block:")
print(out_content)
````

````python
# Nested braces handled correctly
response = """
OUT{
    config: {"nested": {"key": "value"}},
    data: [1, 2, 3]
}
"""

out_content = extract_out_block(response)
print("Nested braces preserved:")
print(out_content)
````

````python
# Strings with braces ignored during scanning
response = """
OUT{
    message: "This {is} a {string} with braces",
    code: "function() { return {}; }"
}
"""

out_content = extract_out_block(response)
print("Braces in strings ignored:")
print(out_content)
````

````python
# OUT block in lndl code fence
response = """
Here's the output:

```lndl
OUT{
    title: t,
    summary: s
}
```
"""

out_content = extract_out_block(response)
print("Extracted from code fence:")
print(out_content)
````

````python
# Missing OUT block raises error
response = "Just some text without an OUT block"

try:
    extract_out_block(response)
except MissingOutBlockError as e:
    print(f"Error: {e}")
````

````python
# Unbalanced braces detected
response = "OUT{ field: value"  # Missing closing brace

try:
    extract_out_block(response)
except MissingOutBlockError as e:
    print(f"Error: {e}")
````

## 6. Parsing OUT{} Block Arrays

OUT blocks support array syntax for multiple variables per field: `field:[var1, var2]`

````python
# Array syntax
out_content = """
title: t,
tags: [tag1, tag2, tag3],
author: a
"""

fields = parse_out_block_array(out_content)
print("Parsed fields:")
for field_name, value in fields.items():
    print(f"  {field_name}: {value} (type: {type(value).__name__})")
````

````python
# Literal values detected
out_content = """
title: "Literal String",
count: 42,
ratio: 3.14,
enabled: true,
disabled: false,
empty: null
"""

fields = parse_out_block_array(out_content)
print("Literal values:")
for field_name, value in fields.items():
    print(f"  {field_name}: {value!r} (type: {type(value).__name__})")
````

````python
# Variable references wrapped in lists
out_content = "title: report_title, author: author_name"

fields = parse_out_block_array(out_content)
print("Variable references:")
for field_name, value in fields.items():
    print(f"  {field_name}: {value} (wrapped in list for consistency)")
````

## 7. Parsing Values

Convert string representations to Python objects: numbers, booleans, lists, dicts.

````python
# Numbers
print("Integer:", parse_value("42"), type(parse_value("42")))
print("Float:", parse_value("3.14"), type(parse_value("3.14")))
print("Negative:", parse_value("-10"), type(parse_value("-10")))
````

````python
# Booleans and None
print("True:", parse_value("true"), type(parse_value("true")))
print("False:", parse_value("false"), type(parse_value("false")))
print("Null:", parse_value("null"), type(parse_value("null")))

# Case insensitive
print("TRUE:", parse_value("TRUE"))
print("False:", parse_value("False"))
````

````python
# Lists and dicts
print("List:", parse_value("[1, 2, 3]"), type(parse_value("[1, 2, 3]")))
print("Dict:", parse_value('{"key": "value"}'), type(parse_value('{"key": "value"}')))
print("Nested:", parse_value('[{"a": 1}, {"b": 2}]'))
````

````python
# Strings (fallback)
print("String:", parse_value("hello world"), type(parse_value("hello world")))
print("Mixed:", parse_value("test-123-abc"))
````

## 8. Full LNDL Workflow Example

Complete example showing variable extraction, action declaration, and OUT block parsing.

````python
# Simulated LLM response with LNDL markup
llm_response = """
I'll analyze the document and provide a structured report.

<lvar Report.title t>Annual Performance Analysis 2024</lvar>
<lvar Report.executive_summary es>The company achieved record revenue of $1.2B,
representing 18% year-over-year growth. Key drivers included product innovation
and market expansion.</lvar>

<lact Report.keywords k>extract_keywords(text=es, count=5, min_score=0.8)</lact>
<lact Report.sentiment sent>analyze_sentiment(text=es)</lact>
<lact validate>validate_report(title=t, summary=es)</lact>

```lndl
OUT{
    title: t,
    executive_summary: es,
    keywords: k,
    sentiment: sent,
    status: "draft"
}
```

Note: The validation action is declared but not referenced in OUT{}, so it won't execute.
"""

print("=== LNDL Response Parsing ===")
print()
````

````python
# Step 1: Extract lvars
lvars = extract_lvars_prefixed(llm_response)
print(f"1. Extracted {len(lvars)} lvars:")
for name, meta in lvars.items():
    value_preview = meta.value[:60] + "..." if len(meta.value) > 60 else meta.value
    print(f"   {name} ({meta.model}.{meta.field}): {value_preview}")
print()
````

````python
# Step 2: Extract lacts
lacts = extract_lacts_prefixed(llm_response)
print(f"2. Extracted {len(lacts)} action declarations:")
for name, meta in lacts.items():
    namespace = f"{meta.model}.{meta.field}" if meta.model else "(direct)"
    print(f"   {name} ({namespace}): {meta.call}")
print()
````

````python
# Step 3: Extract OUT block
out_content = extract_out_block(llm_response)
print("3. Extracted OUT{} block:")
print(out_content)
print()
````

````python
# Step 4: Parse OUT block
fields = parse_out_block_array(out_content)
print("4. Parsed OUT{} fields:")
for field_name, value in fields.items():
    value_type = (
        "literal" if isinstance(value, str) and not isinstance(value, list) else "variable(s)"
    )
    print(f"   {field_name}: {value} ({value_type})")
print()
````

````python
# Step 5: Identify actions to execute
actions_to_execute = set()
for _field_name, value in fields.items():
    if isinstance(value, list):
        for var_name in value:
            if var_name in lacts:
                actions_to_execute.add(var_name)

print("5. Actions to execute (referenced in OUT{}):")
for action_name in actions_to_execute:
    meta = lacts[action_name]
    print(f"   {action_name}: {meta.call}")

not_executed = set(lacts.keys()) - actions_to_execute
if not_executed:
    print(f"\n   Not executed (not referenced in OUT): {', '.join(not_executed)}")
````

## 11. Advanced: Direct Lexer/Parser/AST API

The functions above (`extract_lvars`, `extract_lacts`, etc.) are convenience utilities built on top of a more powerful Lexer/Parser/AST architecture introduced in v1.0.0-alpha4 (PR #194).

**When to use the direct Lexer/Parser/AST API:**
- **Custom validation**: Access token positions for detailed error messages
- **AST manipulation**: Transform or analyze the parse tree before execution
- **Performance-critical paths**: Skip unnecessary parsing when you only need structure
- **Advanced debugging**: Inspect tokenization and parsing steps separately
- **Integration**: Build custom tooling that operates on LNDL AST

**When to use high-level functions** (`parse_lndl`, `extract_*`):
- Standard LNDL parsing with Spec/Operable validation
- Quick extraction without AST manipulation
- Most application code (simpler API)

**Architecture Overview:**
```
LNDL Response → Lexer → Tokens → Parser → AST → Resolver → LNDLOutput
                 ^^^^^^^^^^^^^^^^^^^^
                 Direct API (this section)
```

### 11.1 Basic Lexer Usage

The Lexer performs tokenization, converting LNDL text into a stream of tokens with position tracking (line/column).

````python
# Example LNDL response
text = """
<lvar Report.title t>AI Safety Analysis</lvar>
<lact Report.summary s>generate_summary(text=t, max_words=100)</lact>

OUT{
    title: [t],
    summary: [s],
    confidence: 0.95
}
"""

# Tokenize
lexer = Lexer(text)
tokens = lexer.tokenize()

# Show tokens
print(f"Tokenized into {len(tokens)} tokens:")
print("\nFirst 20 tokens:")
for i, token in enumerate(tokens[:20]):
    print(f"  {i:2d}. {token.type.name:12s} | {token.value!r:20s} | L{token.line}:C{token.column}")

print(f"\n... ({len(tokens) - 20} more tokens)")
print(f"\nLast token: {tokens[-1].type.name} (EOF marker)")
````

````python
# Inspect specific token types
from lionpride.lndl.lexer import TokenType

# Find all LVAR_OPEN tokens
lvar_tokens = [t for t in tokens if t.type == TokenType.LVAR_OPEN]
print(f"Found {len(lvar_tokens)} LVAR_OPEN token(s)")

# Find all OUT_OPEN tokens
out_tokens = [t for t in tokens if t.type == TokenType.OUT_OPEN]
print(f"Found {len(out_tokens)} OUT_OPEN token(s)")

# Find all ID tokens (identifiers)
id_tokens = [t for t in tokens if t.type == TokenType.ID]
print(f"\nIdentifiers found: {[t.value for t in id_tokens[:10]]}")

# Position tracking example
print(
    f"\nFirst identifier '{id_tokens[0].value}' at line {id_tokens[0].line}, column {id_tokens[0].column}"
)
````

### 11.2 Parser Usage

The Parser transforms token streams into Abstract Syntax Trees (AST), providing structured access to LNDL declarations.

````python
# Parse tokens into AST
parser = Parser(tokens, source_text=text)
program = parser.parse()

# Inspect the AST
print("Parsed Program:")
print(f"  Lvars: {len(program.lvars)}")
print(f"  Lacts: {len(program.lacts)}")
print(f"  OUT block: {'Yes' if program.out_block else 'No'}")

# Type information
print("\nAST node types:")
print(f"  Program type: {type(program).__name__}")
if program.lvars:
    print(f"  First lvar type: {type(program.lvars[0]).__name__}")
if program.out_block:
    print(f"  OUT block type: {type(program.out_block).__name__}")
````

### 11.3 AST Inspection

The AST provides direct access to parsed declarations with full type safety.

````python
# Access lvar declarations
print("Lvar declarations:")
for lvar in program.lvars:
    if isinstance(lvar, Lvar):
        # Namespaced lvar
        print(f"  {lvar.alias}: {lvar.model}.{lvar.field} = {lvar.content[:50]}")
    elif isinstance(lvar, RLvar):
        # Raw lvar
        print(f"  {lvar.alias}: (raw) = {lvar.content[:50]}")
````

````python
# Access lact declarations
print("Lact declarations:")
for lact in program.lacts:
    namespace = f"{lact.model}.{lact.field}" if lact.model else "(direct)"
    print(f"  {lact.alias}: {namespace}")
    print(f"    Call: {lact.call}")
````

````python
# Access OUT block fields
if program.out_block:
    print("OUT block fields:")
    for field_name, value in program.out_block.fields.items():
        value_type = type(value).__name__
        print(f"  {field_name}: {value!r} ({value_type})")

    # Check for specific fields
    print(f"\nHas 'title' field: {'title' in program.out_block.fields}")
    print(f"Title value: {program.out_block.fields.get('title')}")
````

### 11.4 Comparison: High-level vs Low-level API

Here's when to use each approach:

````python
# HIGH-LEVEL API (convenience functions)
print("=== HIGH-LEVEL API ===")
print("Use when: Quick extraction, standard parsing, simple workflows\n")

# Using compatibility functions defined in cell 1
lvars_dict = extract_lvars_prefixed(text)
print(f"Extracted {len(lvars_dict)} lvars (dict)")
print(f"Access by alias: lvars_dict['t'].value = {lvars_dict['t'].value[:30]}...")

print("\n" + "=" * 50 + "\n")

# LOW-LEVEL API (Lexer/Parser/AST)
print("=== LOW-LEVEL API ===")
print("Use when: AST manipulation, custom validation, position tracking, debugging\n")

lexer = Lexer(text)
tokens = lexer.tokenize()
parser = Parser(tokens, source_text=text)
program = parser.parse()

print(f"Parsed {len(program.lvars)} lvars (AST nodes)")
print(f"Access by index: program.lvars[0].content = {program.lvars[0].content[:30]}...")
print(f"Type information: {type(program.lvars[0]).__name__} node")
print("Position tracking: Available in tokens (line/column)")
````

````python
# Advanced: Custom AST transformation
print("Advanced use case: Filtering lvars by model\n")

# Filter lvars for specific model
report_lvars = [lv for lv in program.lvars if isinstance(lv, Lvar) and lv.model == "Report"]
print(f"Found {len(report_lvars)} lvars for Report model:")
for lv in report_lvars:
    print(f"  - {lv.field}: {lv.content[:40]}...")

# Build alias → content mapping
alias_map = {lv.alias: lv.content for lv in program.lvars}
print(f"\nAlias map: {list(alias_map.keys())}")

# Resolve references in OUT block
if program.out_block:
    print("\nResolving OUT block references:")
    for field, value in program.out_block.fields.items():
        if isinstance(value, list):
            resolved = [alias_map.get(ref, f"<unresolved: {ref}>") for ref in value]
            print(f"  {field}: {resolved[0][:40] if resolved else '(empty)'}...")
        else:
            print(f"  {field}: {value} (literal)")
````
